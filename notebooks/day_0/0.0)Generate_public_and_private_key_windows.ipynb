# If you have a Windows machine, please follow the instructions below before day 1 so we can get your account properly setup with TSCC


## 1) Download MobaXterm

MobaXterm is a terminal interface program that you will need to access TSCC.  
[Click here](http://download.mobatek.net/1092018073012523/MobaXterm_Installer_v10.9.zip) to download the installer for the latest version (v10.9).

This should save the file `MobaXterm_Installer_v10.9.zip` to your default downloads folder.  
Unzip the file, double click on the installer `MobaXterm_installer_10.9.msi` and follow the instructions to install MobaXterm.

If you are having problems with the installation, download the portable (no install required) version [here](http://download.mobatek.net/1092018073012523/MobaXterm_Portable_v10.9.zip).

## 2) Generate a public key

Open MobaXterm (`MobaXterm.exe` for the installer version or `MobaXterm_Personal_10.9.exe` for the portable version).  
We will need to generate a public key to login to the computing server TSCC.

Click on the "Start local terminal" button.

**Type in the command exactly as it appears below and press ENTER to continue.**  
```bash
ssh-keygen -t rsa
```

You should see the following lines appear on your screen. 
>Generating public/private rsa key pair.  
Enter file in which to save the key (/home/mobaxterm/.ssh/id_rsa):

Press ENTER to accept the default location. It will prompt you to type in a password (the cursor will not move). Do not include the `*` character in the password. Press ENTER when you have typed in your password. It will ask you to confirm your password by typing it again. Remember this password!
>Enter passphrase (empty for no passphrase):  
Enter same passphrase again:  
Your identification has been saved in /home/mobaxterm/.ssh/id_rsa.  
Your public key has been saved in /home/mobaxterm/.ssh/id_rsa.pub.  

Finally it will generate some randomart image.

## 4) Log on to TSCC

Ryan will respond with your TSCC username.

Open up MobaXterm and click on the "Start local terminal" button.

**Type the following command (replace ## with the ucsd-train username number that Ryan gave you) and press ENTER.**

```
ssh ucsd-train##@tscc-login.sdsc.edu
```

You might get a security warning the first time you login. Just select "Yes" to move past it. Once you have logged on, there will be a bunch of text on your screen including "Welcome to TSCC"

To get out of TSCC, type "exit" and press enter.
